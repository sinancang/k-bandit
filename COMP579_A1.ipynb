### Q1: Bernoulli Bandit with k-arms

Write a small simulator for a Bernoulli bandit with k arms. The probability of success pi for each arm i ∈ {1, ...k} should be provided as an input. The bandit should have a function called ”sample” which takes as input the index of an action and provides a reward sample. Recall that a Bernoulli bandit outputs either 1 or 0, drawn from a binomial distribution of parameter pk. Test your code with 3 arms of parameters q∗ = [0.5, 0.5 − δ, 0.5 − 2δ], with δ = 0.1. Generate and save a set of 50 samples for each action. For the test, plot one graph for each action, containing the reward values obtained over the 50 draws, the empirical mean of the values, and the true q∗ for each arm. Each graph will have an x-axis that goes to 50, two horizontal lines (true value and estimated value) and a set of points of value 0 and 1.

```python
import random

from typing import List, Dict, Tuple, Callable
from tqdm import tqdm

import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
plt.style.use('default')
```

```python
class Bandit:
    """
    Bernoulli bandit with k-arms

    Instance Attributes:
        prob_successes: List of probabilities for each action
        k: Integer representing the number of actions
    """
    def __init__(self, prob_success: List[float]) -> None:
        self.prob_success = prob_success
        self.k = len(prob_success)

    def sample(self, action: int) -> int:
        """
        Args:
            action: integer in range [0, k-1]

        Return:
            reward_signal: either 1 or 0
        """
        if action >= self.k:
            raise ValueError(f"Invalid action! Can only choose from {self.k} actions (max allowed = {self.k-1}).")

        X = scipy.stats.bernoulli(self.prob_success[action]).rvs()
        return X
```

```python
probs = [0.5, 0.4, 0.3]
bandit = Bandit(probs)
```

```python
def generate_samples(bandit: Bandit, k: int, n: int) -> List[int]:
    """
    Given a bandit, generates n-samples for the k'th action

    Args:
        bandit: Bandit simulator object
        k: Integer in range [0, k-1] representing the action
        n: Non-negative integer for the number of samples

    Returns:
        samples: List of rewards obtained by taking action k, n times
    """
    samples = []
    for i in range(n):
        samples.append(bandit.sample(k))
    return samples
```

```python
"""
Plot the sample points, empirical mean and true action-value for n=50 for each action k
"""

n = 50

fig = plt.figure(figsize=(16,6))
rows = 1
columns = 3

all_samples = []

for action in range(bandit.k):
    samples = generate_samples(bandit, action, n)
    all_samples.append(samples)

    x = np.linspace(0,n, n)

    fig.add_subplot(rows, columns, action + 1)

    plt.scatter(x, samples, label="Sample", s=15, alpha=.5)
    plt.axhline(y = probs[action], label="True mean", c='r', alpha=.5)
    plt.axhline(y = sum(samples)/n, label="Empirical mean", c='g', linestyle='dotted')

    plt.title(f"k={action}, p={probs[action]}, n={n}, r_t={sum(samples)}")
    plt.ylabel("Total Reward")
    plt.xlabel("Samples")
    plt.legend(loc='upper right', bbox_to_anchor=(0.8, 0.8))

plt.show()
```

### Q2: Action Value Estimation

Code the rule for estimating action values discussed in lecture 2, with a fixed learning rate α, in a function called update, and using the incremental computation of the mean presented in lecture 2, in a function called updateAvg. Using the previous data, plot for each action a graph showing the estimated q value as a function of the number of samples, using averaging as well as α = 0.01 and α = 0.1, and the true value. Each graph should have three curves and a horizontal line.

```python
def update(Q_n: float, R_n: float, lr: float) -> float:
    """
    Incremental update to running average

    Args:
        Q_n = Current estimate of the running average after n-1 draws
        R_n = Reward received on the n'th draw
        lr = Floating point learning rate

    Return:
        Q_n+1 = Updated estimate of the running average after n draws
    """
    return Q_n + lr * (R_n - Q_n)

def updateAvg(Q_n: float, R_n: float, n: int) -> float:
    """
    Incremental update to running average

    Args:
        Q_n = Current estimate of the running average after n-1 draws
        R_n = Reward received on the n'th draw
        n = Number of samples

    Return:
        Q_n+1 = Updated estimate of the running average after n draws
    """
    return Q_n + (1/n) * (R_n - Q_n)
```

```python
"""
Plot the true action-value and estimated action-values.

Action-values are estimated in 3 ways:
1. Sample mean computed at each time step t
2. Memory & computationally efficient incremental averaging with learning rate 0.1 and 0.01
"""

fig = plt.figure(figsize=(16,6))
rows = 1
columns = 3

lrs = [0.01, 0.1]


for action in range(len(all_samples)):
    samples = all_samples[action]
    
    incremental_mean_1 = []
    incremental_mean_2 = []
    averaging_mean = []

    q_n_inc_1 = 0
    q_n_inc_2 = 0
    q_n_avg = 0

    for i in range(len(samples)):
        q_n_inc_1 = update(q_n_inc_1, samples[i], lrs[0])
        incremental_mean_1.append(q_n_inc_1)
        
        q_n_inc_2 = update(q_n_inc_2, samples[i], lrs[1])
        incremental_mean_2.append(q_n_inc_2)
        
        q_n_avg = updateAvg(q_n_avg, samples[i], i+1)
        averaging_mean.append(q_n_avg)

    q_true = probs[action]

    fig.add_subplot(rows, columns, action + 1)

    plt.plot(incremental_mean_1, label=f"Incremental mean 1, lr={lrs[0]}")
    plt.plot(incremental_mean_2, label=f"Incremental mean 2, lr={lrs[1]}")
    plt.plot(averaging_mean, label="Sample mean")
    plt.axhline(y = q_true, label="True mean", c='r', alpha=.5)

    plt.title(f"k={action}, n={n}")
    plt.ylabel("Action-Value")
    plt.xlabel("Samples")
    plt.legend(loc='upper right', bbox_to_anchor=(1, 1))

plt.show()
```

```python
def experiment_av(bandit: Bandit, k: int, n: int, lrs: List[float]) -> np.array:
    """
    Generates n samples using the bandit simulator provided,
    Computes and returns the sample average and incremental averages for each experiment

    Args:
        bandit: Bernoulli Bandit simulator
        k: integer number of action
        n: integer number of samples
        lrs: list of length two of learning rates for the incremental averages

    Returns:
        results: nd.array containing sample average, followed by incremental averages,
                 followed by sample standard error and incremental averages standard error
    """
    incremental_mean_1 = []
    incremental_mean_2 = []
    averaging_mean = []

    q_n_inc_1 = 0
    q_n_inc_2 = 0
    q_n_avg = 0

    for i in range(n):
        sample = bandit.sample(k)
        
        q_n_inc_1 = update(q_n_inc_1, sample, lrs[0])
        incremental_mean_1.append(q_n_inc_1)
        
        q_n_inc_2 = update(q_n_inc_2, sample, lrs[1])
        incremental_mean_2.append(q_n_inc_2)
        
        q_n_avg = updateAvg(q_n_avg, sample, i+1)
        averaging_mean.append(q_n_avg)

    sample_avg = np.mean(averaging_mean)
    sample_std_error = np.std(averaging_mean) / np.sqrt(n)

    incremental_avg_1 = np.mean(incremental_mean_1)
    incremental_std_error_1 = np.std(incremental_mean_1) / np.sqrt(n)

    incremental_avg_2 = np.mean(incremental_mean_2)
    incremental_std_error_2 = np.std(incremental_mean_2) / np.sqrt(n)

    return np.array([sample_avg, incremental_avg_1, incremental_avg_2,
                     sample_std_error, incremental_std_error_1, incremental_std_error_2])
```

```python
def smoothen(list_y):
    list_x = np.linspace(0, len(list_y), len(list_y))
    poly = np.polyfit(list_x,list_y, 5)
    return np.poly1d(poly)(list_x)
```

```python
fig = plt.figure(figsize=(16,20))
rows = 3
columns = 1

lrs = [0.01, 0.1]

for action in range(bandit.k):
    results = np.zeros((6,0))
    for i in tqdm(range(n_experiments)):
        results = np.insert(results, i, experiment_av(bandit, action, n_samples, lrs), axis=1)
        q_true = probs[action]

    fig.add_subplot(rows, columns, action + 1)
    
    xval = np.arange(0, n_experiments, 1)
    plt.errorbar(xval, smoothen(results[0]), yerr=results[3], label = "Sample Averaging")
    plt.errorbar(xval, smoothen(results[1]), yerr=results[4], label = f"Incremental Mean, lr = {lrs[0]}")
    plt.errorbar(xval, smoothen(results[2]), yerr=results[5], label = f"Incremental Mean, lr = {lrs[1]}")
    plt.axhline(y = bandit.prob_success[action], label="True mean", c='r', alpha=.5)

    plt.title(f"Action-Value Estimates: k={action}, experiments={n_experiments}, samples={n_samples}")
    plt.ylabel("Action-Value (Smoothened)")
    plt.xlabel("Experiment")
    plt.legend(loc='upper right', bbox_to_anchor=(1, 1))

plt.show()
```

Clearly, the learning rate of $0.1$ is better, as $0.01$ consistently under-performs. Compared to the average, using $\alpha=0.1$ results in more or less similarly accurate estimations, but with greater standard error. If we wanted to optimize further, we'd look for $\alpha \in (0.01,0.1)$.

### Q4: $\epsilon$-Greedy Algorithm

Code the $\epsilon$-greedy algorithm discussed in class,with averaging updates,with $\epsilon$ provided as an input.
```python
def epsilon_greedy(bandit: Bandit, epsilon: float, lr: float, time_steps: int):
    """
    Epsilon greedy algorithm for reward maximization
    
    Args:
        bandit: Bernoulli Bandit simulator
        epsilon: Float representing the probability of selecting a random action
        lr: Float for learning rate, set to 0.1 by default. 
            If set to 0. it means averaging instead of incremental mean.
        time_steps: Integer for total number of time steps
    """
    q_t = np.zeros(bandit.k)
    
    rewards = np.zeros(time_steps)
    inst_regrets = np.zeros(time_steps)
    all_q_t = np.zeros((time_steps, len(q_t)))

    for t in range(time_steps):
        if scipy.stats.bernoulli(epsilon).rvs():  # Explore
            a_t = random.randint(0, bandit.k - 1)

        else:
            max_arms = np.where(q_t == max(q_t))[0]

            if len(max_arms) == 1: # One max
                a_t = max_arms[0]
                
            else:  # Multiple max (choose uniformly at random)
                a_t = max_arms[random.randint(0, len(max_arms) - 1)]

        r_t = bandit.sample(a_t)
        rewards[t] = r_t
        
        if lr == 0: # Regular averaging
            q_t[a_t] = updateAvg(q_t[a_t], r_t, t+1)

        else: # Incremental averaging
            q_t[a_t] = update(q_t[a_t], r_t, lr)
            
        all_q_t[t] = q_t
        inst_regrets[t] = bandit.prob_success[0] - bandit.prob_success[a_t]

    return (rewards, all_q_t, inst_regrets, np.cumsum(inst_regrets), np.cumsum(rewards))
```

```python
def experiment_generalized(bandit: Bandit, algorithm: Callable, hyper_parameters: Tuple, n_experiments: int, n_time_steps: int) -> Dict:
    results = dict()
    for value1 in hyper_parameters[0]:
        results[value1] = dict()
        for value2 in hyper_parameters[1]:
            print(f"Running for {hyper_parameters[2]}={value1}, {hyper_parameters[3]}={value2}")
            results[value1][value2] = dict()

            n_optimal_estimated = 0

            all_rewards = np.zeros((n_experiments, n_time_steps))
            all_total_rewards = np.zeros((n_experiments, n_time_steps))
            all_inst_regrets = np.zeros((n_experiments, n_time_steps))
            all_total_regrets = np.zeros((n_experiments, n_time_steps))
            all_av_estimates = np.zeros((n_experiments, n_time_steps, bandit.k))

            for i in tqdm(range(n_experiments)):
                experiment = algorithm(bandit, value1, value2, n_time_steps)

                # Store all rewards
                all_rewards[i] = experiment[0]

                # Store all actions
                all_av_estimates[i] = experiment[1]

                # Store all instantaneous regrets
                all_inst_regrets[i] = experiment[2]

                # Store all total regrets
                all_total_regrets[i] = experiment[3]
                                
                # Store all total rewards
                all_total_rewards[i] = experiment[4]

            results[value1][value2]['all_rewards'] = all_rewards
            results[value1][value2]['all_total_rewards'] = all_total_rewards
            results[value1][value2]['all_av_estimates'] = all_av_estimates
            results[value1][value2]['all_inst_regrets'] = all_inst_regrets
            results[value1][value2]['all_total_regrets'] = all_total_regrets
            
    return results
```

### Q6: Upper Confidence Bound Algorithm (UCB)

Write a function that implements the UCB algorithm discussed in lecture 2. Set c = 2. Plot the same graphs as above for $\alpha = 0.1$, $\alpha = 0.01$, $\alpha = 0.001$ and averaging. Explain briefly the behavior you observe.

```python
def UCB(bandit: Bandit, lr: float, c: float, time_steps: int = 1000):
    """
    Upper Confidence Bound algorithm for reward maximization
    
    Args:
        bandit: Bernoulli Bandit simulator
        c: Hyper-parameter determining initial confidence
        time_steps: Integer for total number of time steps
        lr: Float for learning rate, set to 0.1 by default. 
            if set to 0 then use normal averaging
        
    """
    q_t = np.zeros(bandit.k)
    n_t = np.ones(bandit.k)
    
    all_qt = np.zeros((time_steps, len(q_t)))
    rewards = np.zeros(time_steps)
    inst_regrets = np.zeros(time_steps)

    for t in range(time_steps):
        a_t = np.argmax(q_t + c * np.sqrt(np.log(t+1)) / n_t)

        r_t = bandit.sample(a_t)
        rewards[t] = r_t
        
        if lr == 0.: # Regular averaging
            q_t[a_t] = updateAvg(q_t[a_t], r_t, t+1)

        else: # Incremental averaging
            q_t[a_t] = update(q_t[a_t], r_t, lr)

        n_t[a_t] += 1
        
        all_qt[t] = q_t
        inst_regrets[t] = bandit.prob_success[0] - bandit.prob_success[a_t]

    return (rewards, all_qt, inst_regrets, np.cumsum(inst_regrets), np.cumsum(rewards))
```

### Q7: Thompson Sampling

Write a function that implements the Thompson sampling to be discussed in lecture 4. Plot the same graphs as above. Explain briefly the behavior you observe.

```python
def Thompson(bandit: Bandit, b1: float = 1., b2: float = 1., time_steps: int = 1000):
    s_i = np.zeros(bandit.k)
    f_i = np.zeros(bandit.k)
    
    theta_i = np.zeros(bandit.k)
    
    all_qt = np.zeros((time_steps, bandit.k))
    rewards = np.zeros(time_steps)
    inst_regrets = np.zeros(time_steps)

    for t in range(time_steps):
        for arm in range(bandit.k):
            theta_i[arm] = np.random.beta(s_i[arm] + b1, f_i[arm] + b2)

        a_t = np.argmax(theta_i)
        r_t = bandit.sample(a_t)
        
        if r_t == 1:
            s_i[a_t] += 1
        else:
            f_i[a_t] += 1
            
        all_qt[t] = theta_i
        rewards[t] = r_t
        inst_regrets[t] = bandit.prob_success[0] - bandit.prob_success[a_t]
    
    return (rewards, all_qt, inst_regrets, np.cumsum(inst_regrets), np.cumsum(rewards))
```

#### Q9: Non-Stationary Problem

```python
delta = 0.1
init_probs = [0.5, 0.5 - delta, 0.5 - (2* delta)]
final_probs = [0.5, 0.5 + delta, 0.5 + (2* delta)]
```

```python
class Moving_Bandit(Bandit):
    """
    Non-stationary Bernoulli bandit with k-arms

    Instance Attributes:
        prob_successes: List of probabilities for each action
        k: Integer representing the number of actions
        delta: Floating point number to adjust distribution
    """
    def __init__(self, prob_success: List[float], final_probs: List[float]) -> None:
        super().__init__(prob_success)
        self.count = 0

    def sample(self, action: int) -> int:
        """
        Args:
            action: integer in range [0, k-1]

        Return:
            reward_signal: either 1 or 0
        """
        self.count += 1
        if count == 500:
            self.prob_success = final_probs
        return super().sample(action)
```
